```python
import os
import json

# Caminho do diretório
directory_path = r'C:\Mestrado\Materias\2024-02-Sistemas para Processamento Multimidia\seminario\image-segmentation\dataset\trigo'

# Conjunto para armazenar valores únicos de "label"
unique_labels = set()

# Percorrer todas as subpastas e arquivos no diretório
for root, dirs, files in os.walk(directory_path):
    for file in files:
        if file.endswith('.json'):
            file_path = os.path.join(root, file)
            print(f'Processando arquivo: {file_path}')  # Mensagem de loading
            with open(file_path, 'r', encoding='utf-8') as json_file:
                data = json.load(json_file)
                if 'shapes' in data:
                    for shape in data['shapes']:
                        if 'label' in shape:
                            unique_labels.add(shape['label'])

# Imprimir os valores únicos de "label"
print('Valores únicos de "label" encontrados:')
for label in unique_labels:
    print(label)
```
